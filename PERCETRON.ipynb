1. Entrada 
    - função: recebe os sinais de entrada (características, pixels). 
    - número de neurônios: quantidade de sinais de entrada.
    - número de camadas: 1 camada
2. Saída
    - função: fornece a resposta da rede dado as regras de processamento aplicadas desde a camada de entrada.
    - número de neurônios: 
      - classificação binária (2 classes): 1 neurônio
      - classificação não-binária: quantidade de classes
    - número de camadas: 1 camada

## <a name="tipos_de_arquiteturas_de_redes_neurais"></a> Tipos de arquiteturas de redes neurais (topologia)
- Feedfoward
  - fluxo de dados é sempre em uma única direção ("esquerda para direita")
  - neurônios de uma camada enviam sinais apenas para neurônios da camada seguinte. "Neurônio da camada c para c+1"
  - tipos
    - camada simples: não possui camadas intermediárias
    - múltiplas camadas: possui pelo menos 1 camada intermediária  

As Figuras [4](#perceptron_1) e [5](#perceptron_2) ilustram exemplos de redes que possuem arquitetura _feedfoward_. Perceba que as setas que representam as conexões sinápticas apontam sempre para uma única direção (da esquerda para a direita)  

- Recorrente
  - fluxo de dados em várias direções
  - neurônios de uma camada podem enviar sinais tanto para neurônios da camada seguinte, quanto da mesma camada ou camada anterior


  
--- 
<center><b>Explicação Complementar </b></center>

--- 
1. **FeedFoward**  
Não confundir os conceitos dos termos _feedfoward_ e _backpropagation_. O primeiro termo está definindo que a rede possui uma arquitetura na qual o fluxo dos **dados** sempre seguem uma única direção. Já o segundo termo define um **algoritmo de otimização para ajuste dos pesos (aprendizado)** da rede neural. 

2. **Recorrente**  
Realimentação é termo utilizado quando um neurônio envia sinais para quaisquer neurônios pertencentes a camadas anteriores. "Neurônio da camada c para c-1"

## <a name="propagacao_do_sinal"></a> Propagação do sinal
O sinal se propaga na rede de um neurônio para outro através de conexões, as sinapses ponderadas. O neurônio receptor realiza uma combinação linear das saídas de cada neurônio emissor com os pesos das conexões sinápticas. Normalmente, **os pesos sinápticos da rede são inicializados aleatoriamente em uma distribuição normal entre [0, 1]**. Essa combinação linear determina o potencial de ativação do neurônio receptor, e é dada por: 

<a name="mjx-eqn-a"></a>
$$
  \tag{a} 
p(j) = 
\begin{bmatrix}
x_1 & x_2 & ... & x_n
\end{bmatrix}  
\times  
\begin{bmatrix}
w_1 \\
w_2 \\
... \\
w_n
\end{bmatrix}
=
\begin{bmatrix}
x_1w_1 + x_2w_2 + ... x_nw_n 
\end{bmatrix}
$$

$$
ou
$$

<a name="mjx-eqn-b"></a>
$$
\tag{b}
p(j) = \sum_{i = 1}^{n}(x_i \cdot w_i) 
$$


onde:
- $x$ representa o sinal de cada neurônio emissor
- $x_i$ representa o i-ésimo neurônio emissor
- $n$ é o número de neurônios emissores
- $w_i$ representa o peso da conexão entre o i-ésimo neurônio emissor e o neurônio receptor
- $j$ representa o j-ésimo neurônio receptor
- $p(j)$ representa o potencial de ativação do j-ésimo neurônio receptor

Nas Figuras [4](#perceptron_1) e [5](#perceptron_2), os pesos estão associados às conexões sinápticas ilustradas pelas setas em $\color{red}{\text{vermelho}}$. Para generalizarmos a equação acima para qualquer neurônio, primeiro precisamos de mais uma engrenagem matemática do funcionamento de nossa rede neural, o conceito de **Função de ativação**. 



##<a name="funcao_de_ativacao"></a>  Função de ativação
- Definição: é uma função matemática aplicada em todos os neurônios de uma rede neural, exceto nos neurônios da camada de entrada. 
- Objetivo: normalizar o potencial de ativação a uma determinada faixa de valores. Geralmente entre $[-1,1]$ ou $[0,1]$
- Tipos: o tipo pode variar de acordo com o tipo de problema ou com o tipo da camada. Elas podem ser linear, aplicada a problemas linearmente separáveis, ou não-linear. Alguns exemplos de função de ativação são mostrados na Figura [6](#sharma2017).

Agora que temos o conceito de função de ativação, vamos incrementar a nossa equação anterior. Vamos generalizar a equação mostrada na Seção [Propagação do Sinal](#propagacao_do_sinal) para uma rede de $c$ camadas quaisquer, seguindo a notação contida em [(NIELSEN, 2015)](http://neuralnetworksanddeeplearning.com/chap2.html#warm_up_a_fast_matrix-based_approach_to_computing_the_output_from_a_neural_network). Dada uma rede neural com $c$ camadas, cada camada com $n$ neurônios, podemos expandir a Equação [(b)](#mjx-eqn-b) para:  

<a name="mjx-eqn-c"></a>
$$ 
\tag{c}
a^i_j = \sigma(\sum\limits_k (a^{i-1}_k \cdot w^i_{jk}) + b^i_j)
$$

onde:
- $a^{i-1}_k$ representa o valor da ativação do $k$-ésimo neurônio na $(i-1)$-ésima camada, onde $a^1_k$ é o valor da ativação do $k$-ésimo neurônio da camada de entrada da rede neural
- $w^i_{jk}$ é o peso da conexão sináptica entre o $k$-ésimo neurônio da $(i-1)$-ésima camada para o $j$-ésimo neurônio da $i$-ésima camada
- $b^i_j$ é o termo bias do $j$-ésimo neurônio na $i$-ésima camada. O termo bias ainda não foi definido. Definiremos mais adiante nas próximas seções.
- $\sigma$ é a função de ativação (Exemplos na Figura [6](#sharma2017)), e 
- $a^i_j$ é o valor de ativação (saída) do $j$-ésimo neurônio na $i$-ésima camada

## Função custo
Função custo, função de perda ou _loss function_ é um dos principais conceitos de redes neurais relacionado a etapa de aprendizado. Entender o seu objetivo, tipos e quando utilizar cada tipo é de fundamental importância para o projeto e aplicação de redes neurais. O foco desta seção é apresentar ao leitor objetivo e a intuição dessa ferramenta. Os detalhes matemáticos serão brevemente discutidos na próxima seção ([Gradiente Descendente](#gradiente_descendente)) e melhor explorados nos tópicos práticos de redes neurais ([Perceptron](#perceptron), [Multi Layer Perceptron](#mlp)).

Para entendermos melhor a função custo e qual a sua utilidade, vamos adaptar um pouco o exemplo contido em [(MCDONALD, 2017)](https://towardsdatascience.com/machine-learning-fundamentals-via-linear-regression-41a5d11f5220). Imagine que, em uma noite muito fria, uma curiosa criança pequena esteja pela primeira vez próxima ao fogo. O fogo está contido em uma vela. Até o momento, ela não tem conhecimento de que fogo é perigoso, da dor ao tocá-lo e nem a que distância ela pode se aquecer de forma segura. Ela aproxima-se do fogo em 5 momentos da seguinte forma:  
1. a criança curiosa se aproxima do fogo para olhar mais de perto e, percebendo a sensação de calor, se aproxima cada vez mais
2. a criança curiosa toca no fogo contido na vela e corre com medo
3. a criança curiosa não toca no fogo, mas aproxima a sua mão o mais perto sem tocá-lo. Poucos segundos depois, ela sente o ardor do calor e corre com medo
4. a criança curiosa mais uma vez aproxima a mão da vela acesa. Mas, dessa vez, vagarosamente ela distancia a mão até que chegue a uma distância confortável

1. A distância confortável é a distância que mantém a sua mão aquecida, sem sentir dor
  - Ilustrativamente, iremos considerar que essa distância pertence ao intervalo de valores de [0.4, 0.6]. Iremos também nomear os valores 0.4 como $lt$ (do inglês _low threshold_ - menor limiar), 0.6 como $ht$ (do inglês _high threshold_ - maior limiar) e 0.5 como $od$ (do inglês _optimal distance_ - distância ótima)

  - Distâncias acima de $ht$ fornecem dor. Quanto maior, maior a sensação de calor
  - Distâncias entre $lt$ e $ht$ fornecem aquecimento aceitáveis. Mas a distância $od$ fornece o conforto ideal

Através dessa cena ilustrativa, podemos introduzir os seguintes conceitos: 
- Podemos definir feedback como uma punição ou recompensa fornecida pelo ambiente a um sujeito, dada uma ação ou conjunto de ações pertencentes a esse sujeito. 
    - Na nossa cena, a criança recebe punição (frio ou calor) quando não está a uma distância confortável. Caso contrário, recebe recompensa (conforto)
- Podemos definir a experiência do sujeito como a relação entre a ação do sujeito e o feedback fornecido pelo ambiente num instante $t$
    - Na nossa cena, a experiência da criança consiste em ela mover a mão até uma determinada distância $d$ e receber frio, calor ou conforto, num instante $t$
- Podemos definir o processo de aprendizagem como a modificação do conhecimento do sujeito mediante a um conjunto de experiências a que tal sujeito foi submetido.
    - Na nossa cena, após uma experiência, a criança adquire a percepção de que a distância atual deve ser aumentada ou diminuída, mediante ao feedback recebido.

Em nosso exemplo ilustrativo, a criança (rede neural), a partir de um conjunto de experiências (distâncias e feedbacks - amostras anotadas) adquire o conhecimento se deve mover-se ou não (aprendizagem - ajuste dos pesos sinápticos) a partir de uma determinada distância. O quanto a criança deve mover-se para atingir o $od$ é o custo. Assim, podemos definir a **função custo** como uma função ou regra matemática que determina a medida do "quão errado" o modelo neural está a partir de um conjunto de amostras ao qual o modelo foi submetido no processo de aprendizagem. A etapa de aprendizagem (Seção [Processamento Neural](#Processamento-neura)) ocorre com a minimização da magnitude da função custo utilizada e uma direção obtida pelo gradiente.  Esse processo de minimização é regido por algum método de otimização. Em redes neurais, iremos estudar o processo de minimização da função custo através do método **gradiente descendente**.

###Gradiente descendente
Muitas pessoas que utilizam ou estudam redes neurais acreditam que o conceito de gradiente descendente é uma caixa preta, enigmática e difícil entender. Ao invés de ir direto ao formalismo matemático, nesta seção iremos abordar apenas os conceitos básicos, intuição e ilustrar graficamente quais os seus efeitos. 

Vamos adaptar o exemplo contido em [(SHUNG, 2018)](https://towardsdatascience.com/gradient-descent-simply-explained-1d2baa65c757). Imaginemos o seguinte cenário, que iremos chamar de:
___
<a name="cenario_da_montanha"></a> 
[Cenário da Montanha](#cenario_da_montanha)   
1. imagine uma região com várias montanhas e vales.
2. modifique o cenário de tal forma que haja diversos vales, uns mais profundos que outros
    - em nosso cenário hipotético não há abismos, ou seja, não existe descontinuidade  
3. chute 2 números iniciais no intervalo (0.0, 1.0].
4. posicione uma pessoa com os olhos vendados em um ponto qualquer da região que não seja um vale
    - a pessoa não poderá retirar a venda para olhar qual direção seguir. Poderá apenas sentir **a direção da descida**
5. o objetivo é que essa pessoa chegue até **o vale mais profundo** podendo mover-se apenas uma vez por iteração $t$
    - é permitido, como deslocamento, apenas andar ou saltar uma única vez a cada iteração
    - é permitido decidir o quanto se deslocar a cada iteração. Vamos chamar esse "quanto" de taxa de deslocamento $n$
___
  
Montado o cenário e as regras, iremos iniciar o processo de movimentos em direção ao objetivo.
```
---------
Algoritmo
---------
1. Enquanto não chagar a um vale (sentir inclinação 0 (zero) ou próximo)
2. Sinta com um dos pés a direção da descida mais íngreme
3. desloque-se no sentido da descida com taxa de deslocamento $n$
4. retorne ao passo 1
``` 

Estes 4 passos simples resumem não apenas a intuição do que é o GD ou descida mais íngreme, mas também de vários outros conceitos. Segundo [(HAYKIN, 2007)](https://books.google.com.br/books?id=bhMwDwAAQBAJ&dq=HAYKIN+2007&hl=pt-BR&source=gbs_navlinks_s), o método GD consiste em ajustes sucessivos aplicados aos pesos da rede na direção oposta ao gradiente. Vamos relacionar alguns dos conceitos que vimos até agora sobre redes neurais com o cenário da montanha. Mas, primeiramente, vamos resumir o processamento neural com os conceitos que vimos até o momento.  
```
```
Sabemos que os pesos são inicializados aleatoriamente (Seção [Propagação do sinal](#propagacao-do-sinal)). Sabemos também que em redes **_feedfoward_** o sinal de entrada é processado por 2 ou mais camadas de neurônios num fluxo unidirecional através de combinações lineares das multiplicações dos pesos sinápticos com o sinal (Seções [Arquiteturas](#tipos_de_arquiteturas_de_redes_neurais) e [Camadas](#camadas_de_uma_rede_neural)). O resultado em cada neurônio é processado por uma função de ativação em camadas intermediárias e uma função custo na camada de saída. Em cada neurônio da camada de saída, a função custo gera uma probabilidade, que então é comparada com um $target$ (resposta anotada) gerando um sinal de erro. Esse sinal de erro, denominado custo, será utilizado para ajuste dos pesos sinápticos (Seções [Função Custo](#funcao_custo) e [Função de Ativação](#funcao_de_ativação)). O que buscamos é justamente minimizar o erro gerado pela função custo.

No ítem 5 do nosso [Cenário da Montanha](#cenario_da_montanha), mencionamos uma taxa de deslocamento $n$, que controla "a largura do passo" que a pessoa dá em cada momento $t$ em direção à descida mais íngreme. Em redes neurais, essa taxa é chamada de **taxa de aprendizado**, cujo objetivo é controlar a influência do erro, ou melhor, do negativo do gradiente gerado após a aplicação da derivada da função custo sobre o erro. 

###Taxa de aprendizado

Quando treinamos o modelo de aprendizado de máquina, estamos otimizando uma função de custo $J(\theta)$ e esperamos que minimizar seu valor melhore a métrica que nos interessa. A escolha da função custo geralmente está relacionada ao problema específico que pretendemos resolver. Essencialmente, é deliberadamente projetado para indicar a que distância estamos de uma solução ideal. [(SKALSKI, 2019)](#https://towardsdatascience.com/how-to-train-neural-network-faster-with-optimizers-d297730b3713).

Ao utilizar o método GD, podemos nos deparar com 3 problemas principais.
1. Mínimos locais
    - são os pontos onde a função recebe um valor mínimo relativo apenas numa determinada região. Veja a ilustração na Figura [14](#skalski2019_gd_2).

1. Pontos de cela
    - são os platôs, onde o valor da função custo é quase constante. Nesses pontos, o gradiente é quase zerado em todas as direções, impossibilitando fuga. Veja a ilustração na Figura [14](#skalski2019_gd_2).

- Os problemas de mínimo local e ponto de cela pode ser resolvido de 3 formas básicas
  - Instanciar a rede diversas vezes. Uma vez que a inicialização dos pesos da rede é realizada de forma aleatória, o ponto inicial definido pela função custo será diferente a cada instância da rede. Para entender melhor, veja novamente a Figura [10](#gfycat2019_gradient_1) e imagine a ilustração iniciar no outro cume do gráfico. Provavelmente será atingido o outro mínimo da função.
  - Outros otimizadores mais modernos podem ser utilizados, como **_adam_**, **_RMSProp_** e **_nasterov_**.
  - Utilizar o termo **momentun**, que não será discutido nesses tópicos iniciais



3. Explosão do gradiente
    - pode ocorrer em regiões onde a função de custo é muito íngreme. Em tais lugares, o valor do gradiente aumenta drasticamente - o gradiente explosivo - o que leva a grandes passos, muitas vezes arruinando toda a otimização anterior. O problema da explosão do gradiente pode ser evitado pelo recorte ou limitação da influência do gradiente - definindo o valor máximo permitido do gradiente. Isso pode ser realizado através da taxa de aprendizado, que nada mais é que um termo contido no intervalo [0,1] que pondera o gradiente. Definir valores muito altos requer menos iterações, mas pode fazer com que a rede não consiga convergir. Já definir valores pequenos demais necessitará de mais iterações para convergir e irá resolver o problema da explosão do gradiente. No entanto, aumenta a probablidade de não sair de um minimo local ou pontos de cela. As soluções para mínimos locais e para pontos de cela também aplicam-se a este problema.

# Perceptron em prática - problema do and e or
Agora vamos a melhor parte, a parte prática. O perceptron é uma rede feedfoward formada apenas por camadas de entrada e saída
- Objetivo: Resolver o problema do AND e do OR
- Passos:
    1. importar lbs
    2. definir variáveis básicas
    3. montar a rede
    4. preparar os plots
    5. definir a animação e executar



````
# PASSO 1.
%matplotlib inline
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from matplotlib import animation, rc
from IPython.display import HTML

# PASSO 2.
NUM_FEATURES = 2
NUM_ITER = 500
learning_rate = 0.1

# PASSO 3.
# -------------------------
# Esquema da Perceptron
# -------------------------
#
# x0 o----
#         \w    b
#          -----o (sigmoid)
#         /w
# x1 o----
#
#
# -------------------------
# 
# - 4 amostras (possibilidades de 0 e 1) e 2 neurônios na camada de entrada
x = np.array([[0, 0], [1, 0], [1, 1], [0, 1]], np.float32) # 4x2, input

# - resposta (target para )
#y = np.array([0, 0, 1, 0], np.float32) # 4, correct output, AND operation
y = np.array([0, 1, 1, 1], np.float32) # OR operation
 
# - Pesos e bias
W = np.zeros(NUM_FEATURES, np.float32) # 2x1, weight
b = np.zeros(1, np.float32) # 1x1
 
N, d = np.shape(x) # number of samples and number of features
````

````
# PASSO 4.
##################
##################
##################
fig, ax = plt.subplots()
plot_x = np.array([np.min(x[:, 0]), np.max(x[:, 1])])
plot_y = np.array([0, 0], np.float32);
line, = ax.plot(plot_x, plot_y, color='b', linewidth=2) # Returns a tuple of line objects, thus the comma
plt.scatter(x[:, 0], x[:, 1], c=y, s=100, cmap='viridis')
plt.xlim([-0.2, 1.2]); plt.ylim([-0.2, 1.25]);
````

````
# PASSO 5.
def animate(k):
  
  
    global W, b, plot_x, plot_y, x, y, N, learning_rate
    #print(k)    
    for j in range(N):
      #print (j)
      #print (learning_rate)
      #print (plot_x)
      
      # Seção Propagação do Sinal
      p_j = x[j, :].dot(W) + b 
      
      # Seção Função de Ativação - sigmoid
      a_j = 1.0 / (1.0 + np.exp(-p_j)) 
 
      # Seção gradiente 
      cost = y[j] - a_j # error term
      deltaW = cost * x[j, :]
      deltaB = cost
      W = W + learning_rate * deltaW # if err = y - a, then W = W + lRate * deltW
      b = b + learning_rate * deltaB
      
      
    plot_x = np.array([np.min(x[:, 0] - 0.2), np.max(x[:, 1]+0.2)])
    plot_y = - 1 / W[1] * (W[0] * plot_x + b) 
    line.set_ydata(plot_y)
    line.set_xdata(plot_x)
    
    return line,


# Init only required for blitting to give a clean slate.
def init():
    line.set_ydata(plot_y)
    return line,
  
##################
##################
##################
 
ani = animation.FuncAnimation(fig, animate, frames=np.arange(1, NUM_ITER), init_func=init, interval=1000, blit=True)
HTML(ani.to_jshtml())

````

Agora iremos executar o código mostrando os resultados da forma convencional. Teste os parâmetros e observe como os pesos são atualizados

````
import matplotlib.pyplot as plt
import numpy as np

NUM_ITER = 2000
learning_rate = 0.01

# others variables
CHART_GAP = 0.2

x = np.array([[0, 0], [1, 0], [1, 1], [0, 1]], np.float32) # 4x2, input
y = np.array([0, 0, 1, 0], np.float32) # 4, correct output, AND operation
#y = np.array([0, 1, 1, 1], np.float32) # OR operation
 
NUM_FEATURES = x[0].size
W = np.zeros(NUM_FEATURES, np.float32) # 2x1, weight
b = np.zeros(1, np.float32) # 1x1
 
N, d = np.shape(x) # number of samples and number of features
````

````




def draw_plot_old(in_samples, in_line_coords_x, in_line_coords_y, in_gap):

    plt.scatter(in_samples[:, 0], in_samples[:, 1], c=y, s=100, cmap='viridis')
    plt.plot(in_line_coords_x, in_line_coords_y, color='b', linewidth=2)
    
    # define the limits of chart
    plt.xlim([0.0 - in_gap, 1.0 + in_gap]) 
    plt.ylim([0.0 - in_gap, 1.0 + in_gap])
    # show chart
    
   

def draw_plot_new(in_fig, in_object, in_data_y):

    in_object.set_ydata(in_data_y)
    in_fig.canvas.flush_events()
    in_fig.canvas.draw()
    
    


plt.xlim([0.0 - CHART_GAP, 1.0 + CHART_GAP])
plt.ylim([0.0 - CHART_GAP, 1.0 + CHART_GAP])
plt.scatter(x[:, 0], x[:, 1], c=y, s=100, cmap='viridis')
plot_x = np.array([np.min(x[:, 0]), np.max(x[:, 1])])
plot_y = np.array([0, 0], np.float32);
plt.show()

for k in range(NUM_ITER):
    for j in range(N):
      yHat_j = x[j, :].dot(W) + b # 1x2, 2x1
      yHat_j = 1.0 / (1.0 + np.exp(-yHat_j))
 
      err = y[j] - yHat_j # error term
      deltaW = err * x[j, :]
      deltaB = err
      W = W + learning_rate * deltaW # if err = y - yHat, then W = W + lRate * deltW
      b = b + learning_rate * deltaB
      
      # Now plot the fitted line. We need only two points to plot the line
      plot_x = np.array([np.min(x[:, 0]), np.max(x[:, 1]+0.2)])
      if W[1] == 0:
        continue
      else:
        plot_y = - 1 / W[1] * (W[0] * plot_x + b) # comes from, w0*x + w1*y + b = 0 then y = (-1/w1) (w0*x + b)
    
  
    
    draw_plot_old(x, plot_x, plot_y, CHART_GAP)
    

    
    print('W:' + str(W))
    print('b:' + str(b))
    print('plot_x: '+ str(plot_x))
    print('plot_y: '+ str(plot_y))
    print('ITER:' + str(k))
    
````

# Multilayer Perceptron em prática - problema do xor

Agora vamos resolver o problema do XOR

````
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation
from keras.optimizers import SGD
import numpy as np

# -------------------------
# Esquema da MLP
# -------------------------
#
#      w  b
# x0 o----o
#          \w    b
#           -----o (sigmoid)
#      w   /w
# x1 o----o
#         b
#
# -------------------------
# 

X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])

model = Sequential()
model.add(Dense(2, input_dim=2))
model.add(Activation('tanh'))
model.add(Dense(1))
model.add(Activation('sigmoid'))

# Resumo do modelo
model.summary();
````

### Construção da rede
Vamos começar com um modelo linear simples novamente, o percetron. Mas dessa vez, vamos adaptá-lo ao problema de dígitos seguindo os seguintes passos:
1. inicializamos o modelo com **Sequential ()**. A primeira camada é um InputLayer que tem 28 * 28 = 784 nós de entrada (um para cada pixel da imagem) e, em seguida, usamos uma camada Flatten para converter dados de imagem em vetores.
1. adicionamos uma camada de saída como uma camada **Densa com 10 unidades**. A camada Densa conecta cada entrada a cada saída com parâmetros de pesos (os pesos sinápticos).

1. Por último, 
    - selecionamos **_crossentropy_** categórico como a função de perda
    - selecionamos gradiente descendente como o otimizador
    - adicionamos precisão à lista de métricas a serem avaliadas e 
    - compile () para criar o modelo. Note que existem várias opções diferentes para o otimizador em Keras que podemos usar em vez de sgd.

O resumo mostra que existem 7850 parâmetros em nosso modelo, já que a matriz de peso é de tamanho 785x10 (não 784, pois há um termo de polarização adicional). Por último, desenhamos um gráfico mais elaborado do modelo

````
perceptron = Sequential()

perceptron.add(InputLayer(input_shape=(28, 28)))
perceptron.add(Flatten())

perceptron.add(Dense(units=10, activation='softmax'))

perceptron.compile(loss='categorical_crossentropy', 
                 optimizer='sgd', 
                 metrics=['accuracy'])
print(perceptron.summary())

SVG(model_to_dot(perceptron, show_shapes=True).create(prog='dot', format='svg'))
````

### Treinamento da rede

Agora estamos prontos para treinar nosso primeiro modelo. Uma época significa passar por todos os dados de treinamento.

Você pode executar o código abaixo várias vezes e ele continuará o processo de treinamento de onde parou. Se você quiser começar do zero, reinicialize o modelo usando o código na Seção [Testando Perceptron](#testando_perceptron). Por último, plotamos os gráficos da função de custo e da acurácia.


````
%%time
epochs = 10 # one epoch takes about 3 seconds

perceptron_history = perceptron.fit(X_train, Y_train, 
                          epochs=epochs, 
                          batch_size=32,
                          verbose=2)


plt.figure(figsize=(5,3))
plt.plot(perceptron_history.epoch, perceptron_history.history['loss'])
plt.title('loss')

plt.figure(figsize=(5,3))
plt.plot(perceptron_history.epoch, perceptron_history.history['acc'])
plt.title('accuracy');
````

### Teste da rede

Para uma melhor medição da qualidade do modelo, vamos ver a precisão do modelo para os dados de teste.
Podemos agora dar uma olhada nos resultados. Vamos definir uma função auxiliar para mostrar os casos de falha do nosso classificador. 

````
def show_failures(predictions, trueclass=None, predictedclass=None, maxtoshow=10):
    rounded = np.argmax(predictions, axis=1)
    errors = rounded!=y_test
    print('Showing max', maxtoshow, 'first failures. '
          'The predicted class is shown first and the correct class in parenthesis.')
    ii = 0
    plt.figure(figsize=(maxtoshow, 1))
    for i in range(X_test.shape[0]):
        if ii>=maxtoshow:
            break
        if errors[i]:
            if trueclass is not None and y_test[i] != trueclass:
                continue
            if predictedclass is not None and rounded[i] != predictedclass:
                continue
            plt.subplot(1, maxtoshow, ii+1)
            plt.axis('off')
            plt.imshow(X_test[i,:,:], cmap="gray")
            plt.title("%d (%d)" % (rounded[i], y_test[i]))
            ii = ii + 1
````

Aqui estão os primeiros 10 dígitos de teste do modelo linear classificado para uma classe errada:

````
percetron_predictions = perceptron.predict(X_test)
show_failures(percetron_predictions)
````

Queremos que a nossa rede funcione conforme a Figura [20](#gfycat2019_mlp_minist_3), mas vamos utilizar apenas uma camada escondida. Sinta-se a vontade de modificar o código para 2 ou 3.

Dropout ajusta aleatoriamente uma fração de entradas para zero durante o treinamento, que é uma abordagem para a regularização e às vezes pode ajudar a evitar overfitting.

Utilizaremos como função de ativação da camada de saída a função softmax
Finalmente, novamente compilamos () o modelo, dessa vez usando o RMSProp como otimizador.

````
# Model initialization:
model = Sequential()
model.add(InputLayer(input_shape=(28, 28)))
model.add(Flatten())

# A simple model:
model.add(Dense(units=20))
model.add(Activation('relu'))

# A bit more complex model:
#model.add(Dense(units=50))
#model.add(Activation('relu'))
#model.add(Dropout(0.2))

#model.add(Dense(units=50))
#model.add(Activation('relu'))
#model.add(Dropout(0.2))

# The last layer needs to be like this:
model.add(Dense(units=10, activation='softmax'))

model.compile(loss='categorical_crossentropy', 
              optimizer='rmsprop', 
              metrics=['accuracy'])
print(model.summary())

SVG(model_to_dot(model, show_shapes=True).create(prog='dot', format='svg'))

%time
epochs = 10 # one epoch with simple model takes about 4 seconds

history = model.fit(X_train, Y_train, 
                    epochs=epochs, 
                    batch_size=32,
                    verbose=2)

plt.figure(figsize=(5,3))
plt.plot(history.epoch,history.history['loss'])
plt.title('loss')

plt.figure(figsize=(5,3))
plt.plot(history.epoch,history.history['acc'])
plt.title('accuracy');
````

````
%%time
scores = model.evaluate(X_test, Y_test, verbose=2)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))

predictions = model.predict(X_test)

show_failures(predictions)

show_failures(predictions, trueclass=6)
````

## Extra - Montando a matriz de confusão


Também podemos calcular a matriz de confusão para ver quais dígitos são mais misturados e examinar as precisões de classificação separadamente para cada classe:

````
from sklearn.metrics import confusion_matrix

print('Confusion matrix (rows: true classes; columns: predicted classes):'); print()
cm=confusion_matrix(y_test, np.argmax(predictions, axis=1), labels=list(range(10)))
print(cm); print()

print('Classification accuracy for each class:'); print()
for i,j in enumerate(cm.diagonal()/cm.sum(axis=1)): print("%d: %.4f" % (i,j))
````
