1. Entrada 
    - função: recebe os sinais de entrada (características, pixels). 
    - número de neurônios: quantidade de sinais de entrada.
    - número de camadas: 1 camada
2. Saída
    - função: fornece a resposta da rede dado as regras de processamento aplicadas desde a camada de entrada.
    - número de neurônios: 
      - classificação binária (2 classes): 1 neurônio
      - classificação não-binária: quantidade de classes
    - número de camadas: 1 camada

## <a name="tipos_de_arquiteturas_de_redes_neurais"></a> Tipos de arquiteturas de redes neurais (topologia)
- Feedfoward
  - fluxo de dados é sempre em uma única direção ("esquerda para direita")
  - neurônios de uma camada enviam sinais apenas para neurônios da camada seguinte. "Neurônio da camada c para c+1"
  - tipos
    - camada simples: não possui camadas intermediárias
    - múltiplas camadas: possui pelo menos 1 camada intermediária  

As Figuras [4](#perceptron_1) e [5](#perceptron_2) ilustram exemplos de redes que possuem arquitetura _feedfoward_. Perceba que as setas que representam as conexões sinápticas apontam sempre para uma única direção (da esquerda para a direita)  

- Recorrente
  - fluxo de dados em várias direções
  - neurônios de uma camada podem enviar sinais tanto para neurônios da camada seguinte, quanto da mesma camada ou camada anterior


  
--- 
<center><b>Explicação Complementar </b></center>

--- 
1. **FeedFoward**  
Não confundir os conceitos dos termos _feedfoward_ e _backpropagation_. O primeiro termo está definindo que a rede possui uma arquitetura na qual o fluxo dos **dados** sempre seguem uma única direção. Já o segundo termo define um **algoritmo de otimização para ajuste dos pesos (aprendizado)** da rede neural. 

2. **Recorrente**  
Realimentação é termo utilizado quando um neurônio envia sinais para quaisquer neurônios pertencentes a camadas anteriores. "Neurônio da camada c para c-1"

## <a name="propagacao_do_sinal"></a> Propagação do sinal
O sinal se propaga na rede de um neurônio para outro através de conexões, as sinapses ponderadas. O neurônio receptor realiza uma combinação linear das saídas de cada neurônio emissor com os pesos das conexões sinápticas. Normalmente, **os pesos sinápticos da rede são inicializados aleatoriamente em uma distribuição normal entre [0, 1]**. Essa combinação linear determina o potencial de ativação do neurônio receptor, e é dada por: 

<a name="mjx-eqn-a"></a>
$$
  \tag{a} 
p(j) = 
\begin{bmatrix}
w_1 \\
w_2 \\
... \\
w_n
\end{bmatrix}
\times  
\begin{bmatrix}
x_1 & x_2 & ... & x_n
\end{bmatrix}  
=
\begin{bmatrix}
w_1x_1 + w_2x_2 + ... w_nx_n 
\end{bmatrix}
$$

$$
ou
$$

<a name="mjx-eqn-b"></a>
$$
\tag{b}
p(j) = \sum_{i = 1}^{n}(w_i \cdot x_i) 
$$


onde:
- $x$ representa o sinal de cada neurônio emissor
- $x_i$ representa o i-ésimo neurônio emissor
- $n$ é o número de neurônios emissores
- $w_i$ representa o peso da conexão entre o i-ésimo neurônio emissor e o neurônio receptor
- $j$ representa o j-ésimo neurônio receptor
- $p(j)$ representa o potencial de ativação do j-ésimo neurônio receptor

Nas Figuras [4](#perceptron_1) e [5](#perceptron_2), os pesos estão associados às conexões sinápticas ilustradas pelas setas em $\color{red}{\text{vermelho}}$. Uma vez que o sinal é propagado por toda a rede, um custo é calculado e utilizado pela rede para ajuste de seus parâmetros. Este custo é calculado pela **Função Custo**, o qual veremos na próxima Seção.

## Função custo
Função custo, função de perda ou _loss function_ é um dos principais conceitos de redes neurais relacionado a etapa de aprendizado. Entender o seu objetivo, tipos e quando utilizar cada tipo é de fundamental importância para o projeto e aplicação de redes neurais. O foco desta seção é apresentar ao leitor objetivo e a intuição dessa ferramenta.

Para entendermos melhor a função custo e qual a sua utilidade, vamos adaptar um pouco o exemplo contido em [(MCDONALD, 2017)](https://towardsdatascience.com/machine-learning-fundamentals-via-linear-regression-41a5d11f5220). Imagine que, em uma noite muito fria, uma curiosa criança pequena esteja pela primeira vez próxima ao fogo. O fogo está contido em uma vela. Até o momento, ela não tem conhecimento de que fogo é perigoso, da dor ao tocá-lo e nem a que distância ela pode se aquecer de forma segura. Ela aproxima-se do fogo em 5 momentos da seguinte forma:  
1. a criança curiosa se aproxima do fogo para olhar mais de perto e, percebendo a sensação de calor, se aproxima cada vez mais
2. a criança curiosa toca no fogo contido na vela e corre com medo
3. a criança curiosa não toca no fogo, mas aproxima a sua mão o mais perto sem tocá-lo. Poucos segundos depois, ela sente o ardor do calor e corre com medo
4. a criança curiosa mais uma vez aproxima a mão da vela acesa. Mas, dessa vez, vagarosamente ela distancia a mão até que chegue a uma distância confortável

Através dessa cena ilustrativa, podemos introduzir os seguintes conceitos: 
- Podemos definir feedback como uma punição ou recompensa fornecida pelo ambiente a um sujeito, dada uma ação ou conjunto de ações pertencentes a esse sujeito. 
    - Na nossa cena, a criança recebe punição (frio ou calor) quando não está a uma distância confortável. Caso contrário, recebe recompensa (conforto)
- Podemos definir a experiência do sujeito como a relação entre a ação do sujeito e o feedback fornecido pelo ambiente num instante $t$
    - Na nossa cena, a experiência da criança consiste em ela mover a mão até uma determinada distância $d$ e receber frio, calor ou conforto, num instante $t$
- Podemos definir o processo de aprendizagem como a modificação do conhecimento do sujeito mediante a um conjunto de experiências a que tal sujeito foi submetido.
    - Na nossa cena, após uma experiência, a criança adquire a percepção de que a distância atual deve ser aumentada ou diminuída, mediante ao feedback recebido.

Em nosso exemplo ilustrativo, a criança (rede neural), a partir de um conjunto de experiências (distâncias e feedbacks - amostras anotadas) adquire o conhecimento se deve mover-se ou não (aprendizagem - ajuste dos pesos sinápticos) a partir de uma determinada distância. O quanto a criança deve mover-se para atingir o $od$ é o custo. Assim, podemos definir a **função custo** como uma função ou regra matemática que determina a medida do "quão errado" o modelo neural está a partir de um conjunto de amostras ao qual o modelo foi submetido no processo de aprendizagem. A etapa de aprendizagem (Seção [Processamento Neural](#processamento-neural)) ocorre com a minimização do gradiente da função custo utilizada. Esse processo de minimização é regido por algum método de otimização. Em redes neurais, iremos estudar o processo de minimização do gradiente da função custo através do método **[gradiente descendente](#gradiente-descendente-0)**.

#### Intuição sobre derivadas

Antes de falarmos em detalhes spbre a função custo, precisamos revisar um pouco sobre derivadas. Sabemos que a derivada nos dá a inclinação da reta tangente num ponto de uma curva definida por uma função. Mas o que realmente isso quer dizer? Qual a intuição por trás e qual a importância? Porque é aplicada? Porque ela fornece a direção da descida mais íngrime? Nesta seção, tentaremos entender de forma intuitiva o que representa a derivada, sem definições ou conceitos formais. Apenas com gráficos e cálculos simples. Espero que todos que lerem esse material consigam absorver a simplicidade deste material. Não se preocupe caso você tenha dificuldades com cálculo. Apenas tente acompanhar e entender.

A Figura abaixo ilustra o gráfico de uma função linear de apenas uma variável, f(a) = 3a. É fato que quando tomamos a=2 temos que f(a) = 6. Agora, nos deslocarmos a partir de a 0.001 unidades, ou seja, se fizermos a' = a + 0.001, teremos que a' = 2 + 0.001 = 2.001. Até aqui tudo bem correto? E o que acontece com f(a) ? Teremos f(a)'' = 3(2+0.001) = 6.003. Perceba que, deslocar 0.001 unidades na horizontal, ou seja, a partir de a, refletiu em deslocar 0.003 em f(a), ou seja, na vertical. Guarde isso:

1. para a função f(a) = 3a e a=2, deslocar 0.001 unidades a partir de a refletiu em adicionar 0.003 em f(a)

Se traçarmos esses deslocamentos, horizontal e vertical, no gráfico, poderemos visualizar um triângulo retângulo, cujos catetos representam tais deslocamentos. A razão entre a altura definida por um dos catetos e a largura definida pelo outro nos dá a taxa de inclinação da função naquele ponto, ou seja, 0.003 / 0.001 = 3, que é justamente o termo independente da função. E de fato se calcularmos a derivada dessa função encontramos:

$$
\text{se } f(a) = 3a \text{, então } \frac{d}{da}3a = 3 
$$

Agora, tomemos a = 5. Neste caso teremos:

$$
\text{para } a = 5 \text{, então } f(a) = 15
$$
$$
\text{para } a' = a + 0.001 \text{, então } a' = 5.001 \text{ e } f(a') = 15.003
$$
$$
\text{inclinação da função}
\begin{cases}
\frac{altura}{largura} = \frac{0.003}{0.001} = 3 \\
\frac{d}{da}f(a) = \frac{d}{da}3a = 3 
\end{cases}
$$

Conclusão final: a derivada nos dá uma função, ironicamente derivada da função original, que define o comportamento do conjunto imagem quando causamos alguma perturbação a algum elemento do domínio da função.

Para ajustar uma reta, precisamos de 2 parâmetros: um coeficiente angular para determinar a inclinação da reta e um coeficiente linear, para determinar onde a reta corta o eixo y. Iremos chamar estes parâmetros de $w_1$ e $w_0$, respectivamente.  Assim, podemos definir a nossa regressão linear como:

<a name="mjx-eqn-c"></a>
$$
\tag{c}
y = w_0 + w_1x
$$

O código abaixo ilustra 3 exemplos nos quais os 2 parâmetros da reta estão ajustados de forma diferentes. Após executar o código, brinque um pouco com os valores de $w_0$ e $w_1$ e verifique o que acontece com as retas.

````
import matplotlib.pyplot as plt
import numpy as np

def linear_regression(w0, w1, x):
    return (w0 + w1*x)

x = np.asarray([0, 1, 2, 3])

w_0 = 1.5
w_1 = 0
plt.subplot(1, 3, 1)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.xlim(0, 3)
plt.ylim(0, 3)
plt.title('$w_0$ = 1.5 e $w_1$ = 0')
plt.plot(x, linear_regression(w0, w1, x))

w_0 = 0
w_1 = 0.5
plt.subplot(1, 3, 2)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.xlim(0, 3)
plt.ylim(0, 3)
plt.title('$w_0$ = 0 e $w_1$ = 0.5')
plt.plot(x, linear_regression(w0, w1, x))

w_0 = 1
w_1 = 0.5
plt.subplot(1, 3, 3)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.xlim(0, 3)
plt.ylim(0, 3)
plt.title('$w_0$ = 1 e $w_1$ = 0.5')
plt.plot(x, linear_regression(w0, w1, x))

plt.tight_layout()
plt.show()
````


Após propagar cada valor x de entrada pela rede, comparamos a resposta de nossa função com o valor resposta-certa. O objetivo é minimizar a distância entre o valor obtido por nossa função e a resposta-certa. Essa distância é obtida pela derivada da função custo. No caso da regressão linear, normalmente é utilizado o erro médio quadrático como função custo. Assim, este processo pode ser definido como:

<a name="mjx-eqn-d"></a>
$$
\tag{d}
cost(w_0, w_1) = minimize_{w_0w_1} (error)
$$


Para um conjunto de $m$ exemplos dos dados de treinamento, podemos definir a nossa função custo da regressão linear como:

<a name="mjx-eqn-e"></a>
$$
\tag{e}
cost(w_0,w_1)=\frac{1}{2m}  \sum_{i=1}^{m}(\hat{y}^{(i)} - y^{(i)})^2
$$

<a name="mjx-eqn-f"></a>
$$
\tag{f}
cost(w_0,w_1)=\frac{1}{2m}  \sum_{i=1}^{m}(w_0 + w_1x^{(i)} - y^{(i)})^2
$$

Precisamos ajustar o valor das variáveis que podemos mexer em nossa função para que o erro minimize. Na equação acima, não podemos mexer na entrada $x$ e nem na resposta-certa $y$, mas podemos mexer nas variáveis $w_0$ e $w_1$. Então vamos derivar a nossa função em termos de derivadas parciais para estas duas variáveis com o objetivo de minimizar o custo. Chamaremos $cost$ de $J$ para simplificar. 

Primeiro, colocamos a Equação [f](#mjx-eqn-f) em termos de derivadas parciais:

<a name="mjx-eqn-g"></a>
$$
\tag{g}
\frac{\partial}{\partial w_0w_1}J({w_0, w_1}) =  \frac{\partial}{\partial w_0w_1}(\frac{1}{2m}  \sum_{i=1}^{m}(w_0 + w_1x^{(i)} - y^{(i)})^2)
$$
Derivando a Equação [g](#mjx-eqn-g) em relação a $w_0$, temos:

<a name="mjx-eqn-h"></a>
$$
\tag{h}
\frac{\partial}{\partial w_0}J({w_0, w_1}) =  \frac{\partial}{\partial w_0}(\frac{1}{2m}  \sum_{i=1}^{m}(w_0 + w_1x^{(i)} - y^{(i)})^2)
$$

Aplicando a regra da potência e regra da cadeia na Equação [h](#mjx-eqn-h):
<a name="mjx-eqn-i"></a>
$$
\tag{i}
\frac{\partial}{\partial w_0}J({w_0, w_1}) = 2* \frac{1}{2m}   \sum_{i=1}^{m}(w_0 + w_1x^{(i)} - y^{(i)}) * (c + 0 - 0)
$$
$$
\frac{\partial}{\partial w_0}J({w_0, w_1}) = \frac{1}{m}   \sum_{i=1}^{m}(\hat{y}^{(i)} - y^{(i)})
$$

Agora vamos derivar a Equação [g](#mjx-eqn-g) em relação a $w_1$ aplicando as mesmas regras:
<a name="mjx-eqn-j"></a>
$$
\tag{j}
\frac{\partial}{\partial w_1}J({w_0, w_1}) =  \frac{\partial}{\partial w_1}(\frac{1}{2m}  \sum_{i=1}^{m}(w_0 + w_1x^{(i)} - y^{(i)})^2)
$$
$$
\frac{\partial}{\partial w_1}J({w_0, w_1}) = 2* \frac{1}{2m}   \sum_{i=1}^{m}(w_0 + w_1x^{(i)} - y^{(i)}) * (0 + 1*x^{(i)} - 0)
$$
$$
\frac{\partial}{\partial w_1}J({w_0, w_1}) = \frac{1}{m}  \sum_{i=1}^{m}(\hat{y}^{(i)} - y^{(i)}) * x_1^{(i)}
$$

Em resumo, se tivermos uma função com mais variáveis e continuar derivando esta função custo para cada uma delas, você verificará que chegará a seguinte fórmula geral:
<a name="mjx-eqn-k"></a>
$$
\tag{k}
\frac{\partial}{\partial w_j}J(w) = \frac{1}{m}  \sum_{i=1}^{m}(\hat{y}^{(i)} - y^{(i)}) * x_j^{(i)}
$$

Para mais informações sobre os conceitos específicos que aplicamos aqui sobre derivadas parciais, você pode consultar os materiais contidos em [(HERMAN AND STRANG, 2020)](https://math.libretexts.org/Bookshelves/Calculus/Book%3A_Calculus_(OpenStax)/14%3A_Differentiation_of_Functions_of_Several_Variables/14.5%3A_The_Chain_Rule_for_Multivariable_Functions#:~:text=In%20single%2Dvariable%20calculus%2C%20we,form%20of%20the%20chain%20rule.) e [(TRAIN, 2020)](https://www.youtube.com/watch?v=jc2IthslyzM)

### Gradiente descendente e regressão linear

Agora que já temos uma intuição sobre o que é o gradiente descendente, vamos ver isto de uma forma um pouco mais formal. O gradiente descendente é um algoritmo de otimização utilizado em redes neurais para ajustar os parâmetros/pesos $W$ na fase de treinamento. A função de perda mede o quão bem as saídas $\hat{y}^{(i)}$ se encontram em cada um dos exemplos de treinamento comparando com a resposta-certa $y^{(i)}$. A função de custo empilha essas comparações e aplica uma média. Assim, é natural que desejemos minimizar esse custo e isso quer dizer encontrar valores para os parâmetros $W$ que minimizem a dada função $J(w)$.

Tomemos uma rede neural com apenas 1 peso e 1 bias. O gráfico gerado pela função custo deve ser uma superfície convexa, no qual  $J(w_0, w_1)$ corresponde a altura de um ponto nessa superfície. Minimizar essa essa função significa "caminhar" sobre a superfície em direção a um ponto de mínimo. O nosso objetivo é alcançar o mínimo global da função, ou seja, o ponto em que a média das funções de perda (ou erros) seja próximo de zero. Graficamente, queremos alcançar a "altura mínima na superfície em que o ponto possa estar". 

Para alcançar o ponto de mínimo da função de forma mais rápida, devemos percorrer a superfície na direção da descida mais íngrime (maior inclinação) em cada iteração. Podemos achar isso de forma precisa utilizando o gradiente da função que é calculado através de derivadas. Em outras palavras, o resultado das iterações do algoritmo gradiente descendente corresponde a função "dar passos" na direção da descida mais íngrime até que um limiar de custo (valor de altura) na superfície seja alcançado.

Por hora, vamos simplificar mais para compreendermos melhor. Tomemos uma rede neural simples com apenas uma varíável $W$ a ser atualizada. Assim o gráfico da função custo corresponderia ao gráfico da Figura:

Calcular a atualização do peso dessa rede seria equivalente a:

<a name="mjx-eqn-l"></a>
$$
\tag{l}
w = w - \alpha\frac{dJ(w)}{dw} 
$$

Recapitulando, porque adicionamos o negativo do gradiente e o que o termo $\alpha$ faz?
1. O gradiente nos dá a direção da inclinação correto? Ou seja, nos dá "para onde a função está indo, da esquerda para direita". Se a função é crescente, então devemos caminhar para a esquerda e, descrescente, devemos caminhar para a direita. Em outras palavras, devemos sempre caminhar na direção da descida e é exatamente o que o negativo do gradiente faz.
2. O termo $\alpha$ é chamado de taxa de aprendizado e controla o passo de descida.

Agora podemos voltar para o cenário principal, o 2D. Usando os conceitos definidos na Seção [Função custo e regressão linear](#funcao-custo-regrssao-linear)), as atualizações dos parâmetros da rede podem ser definidas como:

<a name="mjx-eqn-m"></a>
$$
\tag{m}
w_1 = w_1 - \alpha\frac{\partial J(w_0,w_1)}{\partial w_1} 
$$

$$
w_0 = w_0 - \alpha\frac{\partial J(w_0,w_1)}{\partial w_0} 
$$

onde o termo os termos $\frac{\partial J(w_0,w_1)}{\partial w_1}$ e $\frac{\partial J(w_0,w_1)}{\partial w_0}$ correspondem as derivadas parciais da função custo em relação a $w_1$ e $w_0$ respecitivamente e $\alpha$ a taxa de aprendizado. O processo de atualização ocorre com o cálculo de cada novo $w$, mas a atualização de cada um deles ocorre apenas simultaneamente. Assim, um pseudo algoritmo para atualização dos parâmetros através do método do gradiente descendente seria:


---------
Algoritmo
---------

1. Para cada época { 
   2. Para cada batch { 
      3.   Faça feedfoward do batch
      1.   Calcule o custo e exiba
      1.   Calcule o gradiente do custo
      1.   atualize os pesos simultaneamente usando o gradiente calculado através da fórmula $w_j = w_j - \alpha(\frac{1}{m}  \sum_{i=1}^{m}(\hat{y}^{(i)} - y^{(i)}) * x_j^{(i)})$ \\
   7. }
8. }

### Prática 1 - Regressão linear usando função custo e gradiente
Agora, vamos implementar uma atualização simples de pesos em uma regressão linear para aplicar todos os conceitos abordados.

````
import matplotlib.pyplot as plt
import numpy as np

def linear_regression(w0, w1, X):
    return (w0 + w1*X)

def propagate(w1, w0, X, Y):

  m = X.shape[0]

  ### EQUATION C ###
  yhat = w1 * X + w0 
  ### EQUATION C ###
  
  ### EQUATION E ###
  error = yhat-Y
  cost = np.sum(np.power(error,2))
  ### EQUATION E ###

  ### EQUATION K ###
  dw = np.dot(error, X)/m
  db = np.sum(error)/m
  ### EQUATION K ###

  grads = {"dw": dw,
           "db": db}

  return grads, cost

def optimize(w, b, X, Y, num_iterations, learning_rate, print_cost = False):

  costs = []
    
  for i in range(num_iterations):
        
        
      # Cost and gradient calculation (≈ 1-4 lines of code)
      ### START CODE HERE ### 
      grads, cost = propagate(w, b, X, Y)
      ### END CODE HERE ###
        
      # Retrieve derivatives from grads
      dw = grads["dw"]
      db = grads["db"]
        
      # update rule (≈ 2 lines of code)
      ### EQUATION M ###
      w = w - learning_rate * dw
      b = b - learning_rate * db
      ### EQUATION M ###

      print ("Cost after iteration %i: %f" %(i, cost))


      params = {"w": w,
              "b": b}
    
      grads = {"dw": dw,
             "db": db}

  return params, grads



X = np.asarray([0, 1, 2, 3])
Y = np.asarray([0, 1, 2, 3])
w0 = 1
w1 = 0

plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.xlim(0, 3)
plt.ylim(0, 3)
plt.plot(X, linear_regression(w0, w1, X))
plt.show()

params, grads = optimize(w1, w0, X, Y, num_iterations= 100, learning_rate = 0.2, print_cost = False)

print ("w = " + str(params["w"]))
print ("b = " + str(params["b"]))
print ("dw = " + str(grads["dw"]))
print ("db = " + str(grads["db"]))

plt.plot(X, linear_regression(params["b"], params["w"], X))
plt.show()


````

##<a name="funcao_de_ativacao"></a>  Função de ativação
- Definição: é uma função matemática aplicada em todos os neurônios de uma rede neural, exceto nos neurônios da camada de entrada. 
- Objetivo: normalizar o potencial de ativação a uma determinada faixa de valores. Geralmente entre $[-1,1]$ ou $[0,1]$

### Derivada de funções de ativação

Agora que entendemos um pouco sobre funções de ativação e temos uma breve intuição sobre derivadas, iremos entender um pouco sobre a derivada de algumas funções de ativação comumente utilizadas em redes neurais.

- função sigmoid:  $\sigma(z) =\frac{1}{1 + e^{-z}}$
<a name="mjx-eqn-n"></a>
$$
\tag{n} 
\text{derivada de } \sigma(z) = g(z) = \
\begin{cases}\frac{d}{dx}g(z) = \text{derivada/inclinação de g(z) no pponto z} \\
= \frac{1}{1 + e^{-z}}(1-\frac{1}{1 + e^{-z}}) \\
= g(z)(1-(g(z))
\end{cases}
$$

---
---
- exemplo 1:   
$z = 10, g(z) = 1, \frac{d}{dx}g(z) \approx 1(1-1) \approx 0$
---
- exemplo 2:   
$z = 0, g(z) = 0, \frac{d}{dx}g(z) \approx 0(1-0) \approx 0$
---
- exemplo 3:   
$z = 0, g(z) = 0.5, \frac{d}{dx}g(z) \approx 0.5(1-0.5) \approx 0.25$
---
---
---
---

- função tangente hiperbólica:  $tanh(z) = \frac{e^{z}-e^{-z}}{e^{z} + e^{-z}}$
$$
\text{derivada de } tanh(z) = g(z) = \
\begin{cases}\frac{d}{dx}g(z) = \text{derivada/inclinação de g(z) no ponto z} \\
= 1-(tanh(z))^2 \\
= 1-(g(z))^2
\end{cases}
$$

---
---
- exemplo 1:   
$z = 10, g(z) = 1, \frac{d}{dx}g(z) \approx 1-1^2 \approx 0$
---
- exemplo 2:   
$z = 0, g(z) = 0, \frac{d}{dx}g(z) \approx 1-0^0 \approx 0$
---
- exemplo 3:   
$z = 0, g(z) = 1, \frac{d}{dx}g(z) \approx 1-0^2 \approx 1$
---
---
---
---

- função relu:  $relu(z) = max(0,z)$
$$
\frac{d}{dz}relu(z) = g'(z)\text{  = derivada/inclinação de } relu(z) \\
g'(z) = \begin{cases}
0 \text{ se} \ < 0 \\
1 \text{ se} \ >= 0 
\end{cases}
$$

---
---
---
---

- função leaky relu:  $relu(z) = max(0.01z,z)$
$$
\frac{d}{dz}relu(z) = g'(z)\text{  = derivada/inclinação de } relu(z) \\
g'(z) = \begin{cases}
0.01 \text{ se} \ < 0 \\
1 \text{ se} \ >= 0 
\end{cases}
$$
---
---

Agora que temos o conceito de função de ativação, vamos incrementar a nossa equação anterior. Vamos generalizar a equação mostrada na Seção [Propagação do Sinal](#propagacao_do_sinal) para uma rede de $c$ camadas quaisquer, seguindo a notação contida em [(NIELSEN, 2015)](http://neuralnetworksanddeeplearning.com/chap2.html#warm_up_a_fast_matrix-based_approach_to_computing_the_output_from_a_neural_network). Dada uma rede neural com $c$ camadas, cada camada com $n$ neurônios, podemos expandir a Equação [(b)](#mjx-eqn-b) para:  

<a name="mjx-eqn-o"></a>
$$
\tag{o}
z^i_j = \sum\limits_k ( w^i_{jk}\cdot a^{i-1}_k) + b^i_j
$$

$$ 
a^i_j = \sigma(z^i_j)
$$

onde:
- $a^{i-1}_k$ representa o valor da ativação do $k$-ésimo neurônio na $(i-1)$-ésima camada, onde $a^1_k$ é o valor da ativação do $k$-ésimo neurônio da camada de entrada da rede neural
- $w^i_{jk}$ é o peso da conexão sináptica entre o $k$-ésimo neurônio da $(i-1)$-ésima camada para o $j$-ésimo neurônio da $i$-ésima camada
- $b^i_j$ é o termo bias do $j$-ésimo neurônio na $i$-ésima camada. O termo bias ainda não foi definido. Definiremos mais adiante nas próximas seções
- $\sigma$ é a função de ativação (Exemplos na Figura [6](#sharma2017)) 
- $z^i_j$ é o valor do potencial de ativação do $j$-ésimo neurônio na $i$-ésima camada. Chamamos como $p^i_j$ na Equação [b](#mjx-eqn-b)
- $a^i_j$ é o valor de ativação (saída) do $j$-ésimo neurônio na $i$-ésima camada

Agora vamos visualizar a rede neural gerada acima de um ponto de vista um pouco mais matemático. Podemos visualizar a camada $C^{[0]}$ como possuindo 3 linhas e 1 coluna ($C_{3\times1}^{[0]}$) e a camada $C^{[1]}$ como possuindo 4 linhas e 1 coluna ($C_{4\times1}^{[1]}$). Podemos dizer que os pesos que conectam a camada $C^{[1]}$ a camada $C^{[0]}$ podem ser representados por uma matriz $W^{[1]}$ tal que $W^{[1]} = W^{[1]}_{4\times3}$, ou seja, cada um dos 4 elementos da camada $C^{[1]}$ está conectado a todos os 3 elementos da camada $C^{[0]}$. Sendo assim, podemos dizer que:

$$
a_{3 \times 1}^{[0]} \text{: vetor de entradas ou camada } C^{[0]} \\
W_{4 \times 3}^{[1]} \text{: matriz de pesos da camada } C^{[1]} \\
b_{4 \times 1}^{[1]} \text{: vetor bias da camada } C^{[1]}\\
a_{4 \times 1}^{[1]} \text{: vetor saída da camada } C^{[1]} \text{ e entrada da camada } C^{[2]} \\
$$

Tomando os pesos $W_{1}^{[1]}$ como sendo os pesos que conectam todos os neurônios $C^{[0]}$ ao neurônio $C_1^{[1]}$:

<a name="mjx-eqn-p"></a>
$$
\tag{p}
W_{1}^{[1]} = 
\begin{bmatrix}
w_{11}^{[1]} \\
w_{21}^{[1]} \\
w_{31}^{[1]} \\
\end{bmatrix},
$$

e que :

$$
W_{1}^{[1]^T} = 
\begin{bmatrix}
w_{11}^{[1]} & w_{21}^{[1]} & w_{31}^{[1]} \\
\end{bmatrix},
$$

podemos dizer então que:

<a name="mjx-eqn-q"></a>
$$
\tag{q}
W_{3 \times 4}^{[1]} = 
\begin{bmatrix}
W_{1}^{[1]}  & W_{2}^{[1]} & W_{3}^{[1]} & W_{4}^{[1]}\\
\end{bmatrix} = 
\begin{bmatrix}
w_{11}^{[1]} & w_{12}^{[1]} & w_{13}^{[1]} & w_{14}^{[1]} \\
w_{21}^{[1]} & w_{22}^{[1]} & w_{23}^{[1]} & w_{24}^{[1]} \\
w_{31}^{[1]} & w_{32}^{[1]} & w_{33}^{[1]} & w_{34}^{[1]} \\
\end{bmatrix},
$$

e que :

$$
W_{4 \times 3}^{[1]^T} = 
\begin{bmatrix}
w_{11}^{[1]} & w_{21}^{[1]} & w_{31}^{[1]} \\
w_{12}^{[1]} & w_{22}^{[1]} & w_{32}^{[1]} \\
w_{13}^{[1]} & w_{23}^{[1]} & w_{33}^{[1]} \\
w_{14}^{[1]} & w_{24}^{[1]} & w_{34}^{[1]} \\
\end{bmatrix}
=
\begin{bmatrix}
-- & W_{1}^{[1]^T} & -- \\
-- & W_{2}^{[1]^T} & -- \\
-- & W_{3}^{[1]^T} & -- \\
-- & W_{4}^{[1]^T} & -- \\
\end{bmatrix}
$$

Agora, para calcularmos a saída dos neurônios da camada $C^{[1]}$, ou seja, obtermos o vetor coluna $a^{[1]}$:
<a name="mjx-eqn-r"></a>
$$
\tag{r}
Z^{[1]} = 
\begin{bmatrix}
-- & W_{1}^{[1]^T} & -- \\
-- & W_{2}^{[1]^T} & -- \\
-- & W_{3}^{[1]^T} & -- \\
-- & W_{4}^{[1]^T} & -- \\
\end{bmatrix}
\begin{bmatrix}
a_{1}^{[0]} \\
a_{2}^{[0]} \\
a_{3}^{[0]}  \\
\end{bmatrix}
+
\begin{bmatrix}
b_{1}^{[1]} \\
b_{2}^{[1]} \\
b_{3}^{[1]}  \\
b_{4}^{[1]} 
\end{bmatrix}
\\
Z^{[1]} =
\begin{bmatrix}
W_{1}^{[1]^T}A^{[0]} + b_{1}^{[1]}\\
W_{2}^{[1]^T}A^{[0]} + b_{2}^{[1]}\\
W_{3}^{[1]^T}A^{[0]} + b_{3}^{[1]}\\
W_{4}^{[1]^T}A^{[0]} + b_{4}^{[1]}\\
\end{bmatrix}
=
\begin{bmatrix}
z_{1}^{[1]} \\
z_{2}^{[1]} \\
z_{3}^{[1]}  \\
z_{4}^{[1]}
\end{bmatrix}
$$
e que:
<a name="mjx-eqn-s"></a>
$$
\tag{s}
A^{[1]} = 
\begin{bmatrix}
a_{1}^{[1]} \\
--\\
--  \\
a_{4}^{[1]}
\end{bmatrix}
= \sigma(Z^{[1]})
$$

Podemos escrever de outra forma:

$$
z_{1}^{[1]} = W_{1}^{[1]^T} \times A^{[0]} + b_{1}^{[1]}, a_{1}^{[1]} = \sigma(z_{1}^{[1]}) \\
z_{2}^{[1]} = W_{2}^{[1]^T} \times A^{[0]} + b_{2}^{[1]}, a_{2}^{[1]} = \sigma(z_{2}^{[1]}) \\
z_{3}^{[1]} = W_{3}^{[1]^T} \times A^{[0]} + b_{3}^{[1]}, a_{3}^{[1]} = \sigma(z_{3}^{[1]}) \\
z_{4}^{[1]} = W_{4}^{[1]^T} \times A^{[0]} + b_{4}^{[1]}, a_{4}^{[1]} = \sigma(z_{4}^{[1]}) \\
$$


Podemos extrapolar a mesma idéia para a camada $C^{[2]}$:
$$
W_{1 \times 4}^{[2]} \text{: vetor de pesos da camada } C^{[2]}\\
b_{1 \times 1}^{[2]} \text{: vetor bias da camada } C^{[2]}\\
a_{1 \times 1}^{[2]} \text{: vetor camada (de saída) } C^{[2]} \\
$$

Resumindo, seja a rede neural acima com $C$ camadas. Seja $C^{[0]}$ a camada de entrada e $C-1$ a camada de saída. podemos dizer que $W_{4 \times 3}^{[1]}, b_{4 \times 1}^{[1]}$ correspondem aos parâmetros a serem ajustados na camada $C^{[1]}$ e $W_{1 \times 4}^{[2]}, b_{1 \times 1}^{[2]}$  correpondem aos parâmetros a serem ajustados na camada $C^{[2]}$. 
Assim, vetorizando todas as variáveis, dado um conjunto de entradas $X$, podemons dizer que:
$$
Z^{[1]} = W^{[1]}X + B^{[1]} \\
$$
Mas $X$ pode ser chamado de $A^{[0]}$, que nos dá:
$$
Z^{[1]} = W^{[1]}A^{[0]} + B^{[1]} \\
A^{[1]} = \sigma(Z^{[1]}) \\
Z^{[2]} = W^{[2]}A^{[1]}  + B^{[2]} \\
A^{[2]} = \sigma(Z^{[2]}) \\
$$
Colocando as dimensões das matrizes paramétricas, temos:
$$
Z_{4 \times 1}^{[1]} = W_{4 \times 3}^{[1]}A_{3 \times 1}^{[0]} + B_{4 \times 1}^{[1]} \\
a_{4 \times 1}^{[1]} = \sigma(Z_{4 \times 1}^{[1]}) \\
Z_{1 \times 1}^{[2]} = W_{1 \times 4}^{[2]}A_{4 \times 1}^{[1]}  + B_{1 \times 1}^{[2]} \\
A_{1 \times 1}^{[2]} = \sigma(Z_{1 \times 1}^{[2]}) \\
$$

Generalizando para $C$ camadas podemos escrever:
<a name="mjx-eqn-t"></a>
$$
\tag{t}
Z^{[i]} = W^{[i]}A^{[i-1]} + B^{[i]}
$$

$$ 
A^{[i]} =\sigma(Z^{[i]})
$$

### Regressão logística

Dado um vetor de características X, que poderia estar relacionado a uma image que você queira reconhecer como sendo ou não a de um gato, nós queremos um algoritmo que possa produzir uma previsão que chamaremos de:

$$
\hat{y} = P (y=1|x), \text{ onde  } 0 \leq \hat{y} \leq 1
$$

e $\hat{y}$ é a probabilidade de que $y = 1$ dado as características de entrada $x$. Em outras palavras se $x$ é uma imagem, nós queremos que $\hat{y}$ informe, qual é a chance de que esta imagem seja uma imagem de um gato?

dados que os parâmetros da regressão logística serão: $w \in \mathbb{R}^{n_x}, b \in \mathbb{R}$, ou seja, $w$ é um vetor de dimensões $n_x$ e $b$ um número real. Então, dados uma entrada $X$ e os parâmetros $W$ e $b$ como geramos a saída $\hat{y}$?

$$
Parâmetros: w \in \mathbb{R}^{n_x}, b \in \mathbb{R}
$$

$$
z = w^T x + b, \text{   }z \in \mathbb{R}
$$

$$
Saída: \hat{y} = \sigma(z)
$$

onde $\sigma$ é uma função que mapeia os valores de $z$ para uma função de probabilidade entre $[0, 1]$. Chamamos essa função de função de ativação e, no caso da regressão logística utilizaremos a função sigmoid, que é dada por:


$$
\sigma(z) = \frac{1}{1 + e^{-z}} = 
\begin{cases}
\text{se $z > 0$ e grande,} &  \sigma(z) \approx \frac{1}{1 + 0} = 1 \\
\text{se $z < 0$ e grande,} &  \sigma(z) \approx \frac{1}{1 + big} \approx 0
\end{cases}
$$

A função de perda (loss function) é uma função que mede o quão bem está o nosso algoritmo para **um exemplo** de treinamento. Em outras palavras, dados:

$$
 \hat{y}^{(i)} = \sigma(W^T x^{(i)} + b), \text{onde  }  \sigma(z) = \frac{1}{1 + e^{-z^{(i)}}}
$$

$$
\text{Dado  } \{(x^{(1)}, y^{(1)}), ... , (x^{(m)}, y^{(m)})\}, \text{queremos  } \hat{y}^{(i)} \approx y^{(i)}
$$

onde $i$ representa o i-ésimo exemplo do conjunto de treinamento, a função custo que representa a função de perda dados uma resposta $\hat{y}$ e uma saída $y$ pode ser representada por:

<a name="mjx-eqn-u"></a>
$$
\tag{u}
L (\hat{y}^{(i)}, y^{(i)}) = - (y^{(i)}log \hat{y}^{(i)} + (1 - y^{(i)})log(1-\hat{y}^{(i)}))
$$

sendo:
 
 $$
\begin{cases}
\text{se $y = 1$: } &  L(\hat{y}^{(i)}, y^{(i)}) = -log\hat{y} \\
\text{se $y = 0$: } &  L (\hat{y}^{(i)}, y^{(i)}) = -log(1-y)
\end{cases}
$$

Se $y = 1$ então queremos que $log\hat{y}$ seja o maior possível e, para isso, $\hat{y}$ deverá ser o maior possível. No entanto, a função sigmoid limita os valores de $log\hat{y}$ para 1. Se $y = 0$ então queremos que a função de custo seja menor, e isso implica que nós queremos que $log(1-\hat{y})$ seja o maior possível e, para isso, $\hat{y}$ deverá ser o menor possível. Lembrando que a função sigmoid limita os valores de $log\hat{y}$ entre 0 e 1.

Para estender a mesma equação para $n$ exemplos de treinamento, podemos reescrever a equação da seguinte forma:

<a name="mjx-eqn-w"></a>
$$
\tag{w}
J (w, b) = \frac{1}{m} \sum_{i=1}^{m}L(\hat{y}^{(i)}, y^{(i)}) =  \frac{1}{m}\sum_{i=1}^{m}[y^{(i)}log \hat{y}^{(i)} + (1 - y^{(i)})log(1-\hat{y}^{(i)})]
$$

Essa equação é denominada função custo, que é uma função que mede o quão bem está o nosso algoritmo para **todos os exemplos** de treinamento, ou seja, mede o quão bem os parâmetros $W$ e $b$ estão ajustados em relação ao nosso conjunto de treinamento.

````
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation
from keras.optimizers import SGD
import numpy as np

# -------------------------
# Esquema da MLP
# -------------------------
#
#      w  b
# x0 o----o
#          \w    b
#           -----o (sigmoid)
#      w   /w
# x1 o----o
#         b
#
# -------------------------
# 

X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])

model = Sequential()
model.add(Dense(2, input_dim=2))
model.add(Activation('tanh'))
model.add(Dense(1))
model.add(Activation('sigmoid'))

# Resumo do modelo
model.summary();
````

### Construção da rede
Vamos começar com um modelo linear simples novamente, o percetron. Mas dessa vez, vamos adaptá-lo ao problema de dígitos seguindo os seguintes passos:
1. inicializamos o modelo com **Sequential ()**. A primeira camada é um InputLayer que tem 28 * 28 = 784 nós de entrada (um para cada pixel da imagem) e, em seguida, usamos uma camada Flatten para converter dados de imagem em vetores.
1. adicionamos uma camada de saída como uma camada **Densa com 10 unidades**. A camada Densa conecta cada entrada a cada saída com parâmetros de pesos (os pesos sinápticos).

1. Por último, 
    - selecionamos **_crossentropy_** categórico como a função de perda
    - selecionamos gradiente descendente como o otimizador
    - adicionamos precisão à lista de métricas a serem avaliadas e

Dropout ajusta aleatoriamente uma fração de entradas para zero durante o treinamento, que é uma abordagem para a regularização e às vezes pode ajudar a evitar overfitting.

Utilizaremos como função de ativação da camada de saída a função softmax
Finalmente, novamente compilamos () o modelo, dessa vez usando o RMSProp como otimizador.

````
# Model initialization:
model = Sequential()
model.add(InputLayer(input_shape=(28, 28)))
model.add(Flatten())

# A simple model:
model.add(Dense(units=20))
model.add(Activation('relu'))

# A bit more complex model:
#model.add(Dense(units=50))
#model.add(Activation('relu'))
#model.add(Dropout(0.2))

#model.add(Dense(units=50))
#model.add(Activation('relu'))
#model.add(Dropout(0.2))

# The last layer needs to be like this:
model.add(Dense(units=10, activation='softmax'))

model.compile(loss='categorical_crossentropy', 
              optimizer='rmsprop', 
              metrics=['accuracy'])
print(model.summary())

SVG(model_to_dot(model, show_shapes=True).create(prog='dot', format='svg'))

%time
epochs = 10 # one epoch with simple model takes about 4 seconds

history = model.fit(X_train, Y_train, 
                    epochs=epochs, 
                    batch_size=32,
                    verbose=2)

plt.figure(figsize=(5,3))
plt.plot(history.epoch,history.history['loss'])
plt.title('loss')

plt.figure(figsize=(5,3))
plt.plot(history.epoch,history.history['acc'])
plt.title('accuracy');
````

````
%%time
scores = model.evaluate(X_test, Y_test, verbose=2)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))

predictions = model.predict(X_test)

show_failures(predictions)

show_failures(predictions, trueclass=6)
````
